```python
import json
import pandas as pd
import os
import glob

# Opening JSON file
data_location = os.path.join('data', '*.json')
print(data_location)

filenames = glob.glob(data_location)
print(filenames)

# create empty dataframes for subtopics
dfnames = ['df_lwl', 'df_lkq', 'df_prof','df_fc','df_whlockdown','df_protection','df_blame','df_wistle','df_US']
for x in dfnames: exec(x + ' = pd.DataFrame()')
```

```python
for f in filenames:
    # modify the data structure
    datafile = open(f)
    data = json.load(datafile)
    df = pd.DataFrame(data)
    # expand information of the original post being reposted: normalize the column of dictionaries
    df_repost = pd.json_normalize(df.repost)
    # add prefix to avoid overlap of column names
    df_repost = df_repost.add_prefix('Repost_')
    #  and join the repost df with main df
    df = df.join(df_repost)
    # drop original repost column
    df.drop(columns=['repost'], inplace=True)
    # select posts with keywords    
    df_lwl = pd.concat([df_lwl, df[df['content'].str.contains('李文亮')]])
    df_lkq = pd.concat([df_lkq, df[df['content'].str.contains('李克强')]])
    df_prof = pd.concat([df_prof, df[df['content'].str.contains('支援'or'驰援'or'专家天团')] ])
    df_fc = pd.concat([df_fc, df[df['content'].str.contains('方舱'or '方舱医院' or '火神山'or'雷神山')] ])
    df_whlockdown = pd.concat([df_whlockdown, df[df['content'].str.contains('武汉'and'封城')] ])
    df_protection = pd.concat([df_protection, df[df['content'].str.contains('防护'or'口罩')] ])
    df_blame = pd.concat([df_blame, df[df['content'].str.contains('问责'or '懒政')] ])
    df_wistle = pd.concat([df_wistle, df[df['content'].str.contains('吹哨'or'发哨'or'老子到处说')] ]) 
    df_US = pd.concat([df_US, df[df['content'].str.contains('美国'or'川普'or'特朗普'or '蓬佩奥'or '彭斯')] ]) 
```
